```python
def categorise(row):
    if row['Age'] > 17 and row['Age'] <= 25:
        return '18-25'
    elif row['Age'] >25 and row['Age'] <=35:
        return '26-35'
    elif row['Age'] >35 and row['Age'] <=45:
        return '36-45'
    elif row['Age'] >45 and row['Age'] <=55:
        return '46-55'
    return '56-65'

rd_master['Age_Group'] = rd_master.apply(lambda row: categorise(row), axis=1)

print(rd_master)

```

```python
age_profit = rd_master.groupby('Age_Group')['Profit'].mean()
```

```python
age_profit
```

```python
#Now that I added a column to my data so that I could categorize ages, I will now create 
#a bar graph to describe and compare how age groups affect Profit for the two companies.
grouped_multiple=rd_master.groupby(['Company', 'Age_Group']).agg({'Profit':['mean']})
grouped_multiple.columns=['Profitmean']
grouped_multiple=grouped_multiple.reset_index()
grouped_multiple = grouped_multiple.set_index(['Age_Group', 'Company']).Profitmean
```

```python
grouped_multiple.unstack().plot(kind='bar', stacked=True, title='Age Effects on Profit for Pink and Yellow Cab Companies', color=['pink', 'yellow'])
plt.legend(loc='upper center', ncol=3)
plt.ylabel('Profit')
```

```python
# Population influence the profit for cab riders
grouped_multiple=rd_master.groupby(['Company', 'Population']).agg({'Profit':['mean']})
grouped_multiple.columns=['Profitmean']
grouped_multiple=grouped_multiple.reset_index()
grouped_multiple = grouped_multiple.set_index(['Population', 'Company']).Profitmean

```

```python
grouped_multiple.unstack().plot(kind='bar', stacked=True, title='Population Effects on Profit for Pink and Yellow Cab Companies', color=['pink', 'yellow'])
plt.legend(loc='upper center', ncol=3)
plt.ylabel('Profit')
```

```python
# City influence number of Users
grouped_multiple=rd_master.groupby(['Company', 'City']).agg({'Users':['count']})
grouped_multiple.columns=['Usercount']
grouped_multiple=grouped_multiple.reset_index()
grouped_multiple = grouped_multiple.set_index(['City', 'Company']).Usercount
```

```python
grouped_multiple.unstack().plot(kind='bar', stacked=True, title='Cities Effects on number of Users for Pink and Yellow Cab Companies',  color=['pink', 'yellow'])
plt.legend(loc='upper center', ncol=3)
plt.ylabel('Usercount')
```

```python
# Gender of customer can influence Profit
grouped_multiple=rd_master.groupby(['Company', 'Gender']).agg({'Profit':['mean']})
grouped_multiple.columns=['Profitmean']
grouped_multiple=grouped_multiple.reset_index()
grouped_multiple = grouped_multiple.set_index(['Gender', 'Company']).Profitmean
```

```python
grouped_multiple.unstack().plot(kind='bar', stacked=True, title='Gender of customer can influence Profit', color=['pink', 'yellow'])
plt.legend(loc='upper center', ncol=3)
plt.ylabel('Profit')
```

```python
# Number of users have positive difference in price of cab
grouped_multiple=rd_master.groupby(['Company', 'Users']).agg({'Price Charged':['mean']})
grouped_multiple.columns=['Pricemean']
grouped_multiple=grouped_multiple.reset_index()
grouped_multiple = grouped_multiple.set_index(['Users', 'Company']).Pricemean
```

```python
grouped_multiple.unstack().plot(kind='bar', stacked=True, title='Number of Users have positive difference in price  ', color=['pink', 'yellow'])
plt.legend(loc='upper center', ncol=3)
plt.ylabel('Price Charged')
```

```python
# Higher the population, higher the cab riders
grouped_multiple=rd_master.groupby(['Company', 'Population']).agg({'Transaction ID':['mean']})
grouped_multiple.columns=['Transactionmean']
grouped_multiple=grouped_multiple.reset_index()
grouped_multiple = grouped_multiple.set_index(['Population', 'Company']).Transactionmean
```

```python
grouped_multiple.unstack().plot(kind='bar', stacked=True, title='Higher the population greater the cab riders', color=['pink', 'yellow'])
plt.legend(loc='upper center', ncol=3)
plt.ylabel('Transaction ID')
```
